## 【問題2】100日後以外の米粒の数


```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    
    initial_rice = 1
    list_n_grains  = [initial_rice]
    list_total_grains = [initial_rice]

    for i in range(day-1):
        list_n_grains.append(list_n_grains[-1]*2)
        list_total_grains.append(list_total_grains[-1] + list_n_grains[-1])
        
    return list_n_grains, list_total_grains

day = 100 #関数検証用

list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=day)

print("100日目にもらう米の数：", list_n_grains[-1], "粒")
print("100日後に合計でもらっている米の数：", list_total_grains[-1], "粒")

plt.title("Number of rice to receive",**igfont)
plt.xlabel("日数days",**igfont)
plt.ylabel("米の数Number of rice",**igfont)
plt.plot(list_n_grains, label="per day") 
plt.plot(list_total_grains, label="total") 
plt.legend() 
plt.show()
```
