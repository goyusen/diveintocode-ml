## 各特徴量の欠損値の有無を確認する

```python
import missingno as msno
msno.matrix(df)
```

```python
msno.bar(df)
```

```python
sns.heatmap(df.isna(), cbar=False)
```

**Alley、PoolQCなど大量に欠損値がある特徴量はビジュアル的に確認できる**

### 欠損値が何パーセント含まれるかを確認する。

```python
df_missing = df.isna().sum(axis=0).reset_index()
df_missing .columns = ['', 'Total']
df_missing ['missing_ratio'] = df_missing ['Total'] / df['Id'].count()
df_missing  = df_missing .loc[df_missing ['Total'] > 0].sort_values(by=['missing_ratio'],ascending=False)
print(df_missing.to_string(index=False))
```

### 5つ以上欠損値がある特徴量（列）は削除する。

```python
df_drop = df_missing[df_missing['Total'] >= 5]
dropping_columns_list = list(df_drop.iloc[:, 0]) #5つ以上欠損値がある特徴量の列名をリスト化
df_dropped = df.drop(dropping_columns_list, axis=1)
df_dropped
```

### 欠損値があるサンプル（行）は削除する。

```python
df_dropped_fin = df_dropped.dropna(axis=0) #残っている欠損値のあるサンプル（行）を削除する
df_dropped_fin
```

## 【問題6】分布の確認

- seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
- 目的変数に対して対数変換を行う。
- 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。


[seaborn.distplot — seaborn 0.9.0 documentation](https://seaborn.pydata.org/generated/seaborn.distplot.html)

### seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。

(参考記事)
[pandasで歪度(Skewness)と尖度(Kurtosis)を計算](https://qiita.com/ymmy02/items/36e708d3d9e5e4e6a457)

```python
sns.distplot(df['SalePrice']) #目的変数の分布
```

```python
kurtosis = df.kurt()['SalePrice'] #尖度を算出
skewness = df.skew()['SalePrice'] #歪度を算出
print("尖度: {:.2f}  歪度: {:.2f}".format(kurtosis, skewness))
```

### 目的変数に対して対数変換を行う。

```python
SalePrice_log = df['SalePrice'].apply(np.log) #apply()関数を使って、全てのレコードに対数変換処理を行う
sns.distplot(SalePrice_log)
```

```python
kurtosis_logged = SalePrice_log.kurt()
skewness_logged = SalePrice_log.skew()
print("尖度: {:.2f}  歪度: {:.2f}".format(kurtosis_logged, skewness_logged))
```

### 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。

### ⇒対数変換後の尖度と歪度両方とも大幅に減少し、分布も正規分布に近い形となった。
