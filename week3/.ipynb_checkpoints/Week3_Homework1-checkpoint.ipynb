## フィッシャーのアヤメ

「フィッシャーのアヤメ」という機械学習やデータ分析の世界で有名なデータセットを使い、基本的なデータの扱い方を学んでいきます。


これはカナダのガスペ半島のある牧草地で同じ日に採取された以下の3種類のアヤメ、合計150本のデータです。

[Iris setosa - Wikipedia](https://en.wikipedia.org/wiki/Iris_setosa)

[Iris virgicolor - Wikipedia](https://en.wikipedia.org/wiki/Iris_versicolor)

[Iris virginica - Wikipedia](https://en.wikipedia.org/wiki/Iris_virginica)

setosa、virgiclor、virginicaはここで 目的変数 や ラベル と呼ばれるものに当たります。


1本ごとにsepal_length（がく片の長さ）, sepal_width（がく片の幅）, petal_length（花弁の長さ）, petal_width（花弁の幅）の4種類の情報が記録されています。この4種類を 特徴量 と呼びます。

以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。


[sklearn.datasets.load_iris — scikit-learn 0.20.2 documentation](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html)

## 【問題4】irisデータセット自体の調査

ページのはじめで「カナダのガスペ半島のある牧草地で」と書きましたが、データセット自体についてもう少し調査しておきましょう。データセットを見る上で、そのデータセット自体について知っておくことも大切です。

アヤメの花のデータセットまたはフィッシャーのアヤメのデータセットは、イギリスの統計学者、優生学者、生物学者であるロナルド・フィッシャーが1936年に発表した論文The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis（線形判別分析の一例としての分類学的問題における複数測定値の使用）で紹介した多変量データセットである。 エドガー・アンダーソン（Edgar Anderson）が3つの近縁種のアイリスの花の形態学的変化を定量化するためにデータを収集したことから、アンダーソンのアイリスデータセットと呼ばれることもある。

データセットは、3種のアイリス（Iris setosa, Iris virginica, Iris versicolor）の各50サンプルから構成されています。各サンプルから4つの特徴を測定した：萼と花弁の長さと幅（センチ単位）。これら4つの特徴の組み合わせに基づいて、フィッシャーは、種を互いに区別するための線形判別モデルを開発した。

### 箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。


**箱ひげ図は四分位点や外れ値などは見やすいが、データの分布がどうなっているのかわからない。バイオリン図はデータの分布に応じて箱の形が変わり見やすいが、表示しているのはあくまでも密度であり実際の数の大きさではないため、幅を比較しても数の比較はできない。**

### sns.swarmplot()で密度ではなく、実際の数の分布を確認してみる

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

sns.swarmplot(x='Species', y="sepal length (cm)", data=df, size=3.5, ax=axes[0, 0])
sns.swarmplot(x='Species', y="sepal width (cm)", data=df, size=3.5, ax=axes[0, 1])
sns.swarmplot(x='Species', y="petal length (cm)", data=df, size=3.5, ax=axes[1, 0])
sns.swarmplot(x='Species', y="petal width (cm)", data=df, size=3.5, ax=axes[1, 1])
```

## 【問題7】特徴量間の関係の確認

特徴量間の関係を見ていきます。次の4つを行ってください。


- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る(６つ)
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る(１つ)
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る(１つ)

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。


[pandas.DataFrame.corr — pandas 0.24.2 documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.corr.html)

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る(６つ)

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 10))

x1, x2, x3, x4 = 'petal length (cm)','petal width (cm)','sepal length (cm)','sepal width (cm)'
y1, y2, y3, y4 = 'petal length (cm)','petal width (cm)','sepal length (cm)','sepal width (cm)'

axes[0, 0].scatter(df[df['Species']==0][x1], df[df['Species']==0][y2], color='r', label='Species:0')
axes[0, 0].scatter(df[df['Species']==1][x1], df[df['Species']==1][y2], color='g', label='Species:1')
axes[0, 0].scatter(df[df['Species']==2][x1], df[df['Species']==2][y2], color='b', label='Species:2')
axes[0, 0].legend()

axes[0, 1].scatter(df[df['Species']==0][x2], df[df['Species']==0][y3], color='r', label='Species:0')
axes[0, 1].scatter(df[df['Species']==1][x2], df[df['Species']==1][y3], color='g', label='Species:1')
axes[0, 1].scatter(df[df['Species']==2][x2], df[df['Species']==2][y3], color='b', label='Species:2')
axes[0, 1].legend()

axes[1, 0].scatter(df[df['Species']==0][x3], df[df['Species']==0][y4], color='r', label='Species:0')
axes[1, 0].scatter(df[df['Species']==1][x3], df[df['Species']==1][y4], color='g', label='Species:1')
axes[1, 0].scatter(df[df['Species']==2][x3], df[df['Species']==2][y4], color='b', label='Species:2')
axes[1, 0].legend()

axes[1, 1].scatter(df[df['Species']==0][x4], df[df['Species']==0][y1], color='r', label='Species:0')
axes[1, 1].scatter(df[df['Species']==1][x4], df[df['Species']==1][y1], color='g', label='Species:1')
axes[1, 1].scatter(df[df['Species']==2][x4], df[df['Species']==2][y1], color='b', label='Species:2')
axes[1, 1].legend()

axes[2, 0].scatter(df[df['Species']==0][x1], df[df['Species']==0][y3], color='r', label='Species:0')
axes[2, 0].scatter(df[df['Species']==1][x1], df[df['Species']==1][y3], color='g', label='Species:1')
axes[2, 0].scatter(df[df['Species']==2][x1], df[df['Species']==2][y3], color='b', label='Species:2')
axes[2, 0].legend()

axes[2, 1].scatter(df[df['Species']==0][x2], df[df['Species']==0][y4], color='r', label='Species:0')
axes[2, 1].scatter(df[df['Species']==1][x2], df[df['Species']==1][y4], color='g', label='Species:1')
axes[2, 1].scatter(df[df['Species']==2][x2], df[df['Species']==2][y4], color='b', label='Species:2')
axes[2, 1].legend()
plt.show()
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る(１つ)

sns.pairplot(df, hue='Species', palette='Reds')
```

```python
# ４つの特徴量に関する相関係数行列を作る
corr = df.iloc[:, 0:4].corr()
corr
```

```python
# 相関係数行列のヒートマップを作る(１つ)
sns.heatmap(corr, cmap='coolwarm', annot=True)
```

## 【問題8】可視化したグラフや表の結果の説明

データを探索した時には、結果を述べる（+考察する）ことが大切です。ここまでで作成した表や図から読み取れることや気づきを説明してください。


例えば、散布図行列を見ながら、3種類のアヤメにはどのような違いがあるかを説明してください。

- Species[0]のpetal_wdithとpetal_lengthが他のSpeciesより短いのと、petal_wdithが２cm、petal_lengthが0.5cm以下に集中していることが箱ひげ図やバイオリン図、swarmplot図で確認できた。

- Species[2]petal_wdithとpetal_lengthも他のSpeciesより長いのと、petal_wdithが5cm、petal_lengthが1.5cm以下に集中していることが見えるが、Species[1]の一部とかぶっているため、境界を分けることが難しそうであった。

- 相関係数行列から、petal_wdithとpetal_lengthの相関係数が0.96もあり、強い正の相関と言えよう。sepal_lengthとpetal_lengthが0.87、petal_wdithとsepal_lengthが0.82と、比較的に高い相関係数である。
